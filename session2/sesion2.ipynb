```python
print("Processing")
for i in range(10):
    print(f"Starting for i = {i}")
    for j in range(5):
        if i * j % 2 == 0:
            print(f"{i}, {j}, {i*j}, Even")
        else:
            print(f"{i}, {j}, {i*j}, Odd")
    print(f"Done for i = {i}")
print("Done")
```

```python
def convert_eur_ron(eur_amount, EUR2RON=5):
    ron_amount = EUR2RON * eur_amount
    return ron_amount

eur = 100
ron = convert_eur_ron(eur)
print(f"EUR {eur} = RON {ron}")
ron = convert_eur_ron(eur, 5.12)
print(f"EUR {eur} = RON {ron}")
```

```python
def convert_currency(source_amount, conv_rate=None):
    if conv_rate is None:
        print("Invalid conversion rate")
        return

    target_amount = source_amount * conv_rate
    return target_amount

eur = convert_currency(100)
print(eur)
eur = convert_currency(100, 1/5)
print(eur)
```

```python
def product(n):
    p = 1
    for i in range(1, n+1):
        p = p * i
    return p

p = product(10)
print(f"{p}")
```

```python
def read_input(stop_condition):
    i = int(input())
    if i == stop_condition:
        print("Done.")
        return

    while i != stop_condition:
        print(f"Entered {i}.")
        i = int(input())

    print("Done")
```

```python
read_input(10)
```
